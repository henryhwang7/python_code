```python
menu = """
1. Bulgogi burger = 5 dollars
2. Cheese burger = 4 dollars
3. Chicken burger = 6 dollars
4. Beef burgger = 8 dollars
5. Ice Cream = 2 dollars
7. Completing order
Choose your menu: """

while True:
    money = 0
    menulist = []
    choice = 0
    price = 0
    
    while True:
        print(menu)
        choice = int(input())
        if choice == 7:
            print("\nYou finished your order\n")
            break
        elif choice > 5 and choice !=0 and choice !=7:
            print("Type error. Please, choose the menu")
        else:
            menulist.append(choice)
            print("Your order: {0}\n".format(choice))
            print("Your current order: {0}".format(menulist))
    print("------- Order Review ---------")
    
    while menulist:
        a = menulist.pop()
        if a == 1:
            print("Bulgogi burger order complete")
            price += 5
        elif a == 2:
            print("Cheese burger order complete")
            price += 4
        elif a == 3:
            print("Chicken burger order complete")
            price += 6
        elif a == 4:
            print("Beef burger complete")
            price += 8
        elif a == 5:
            print("Icecream order complete")
            price += 2
        print("Current price: {0} dollars".format(price))
    
    print("Final Price: {0} dollars\n".format(price))
    print("----------- Payment ------------")
    
    while True:
        money = int(input("Please, pay your balance due: "))
        
        if money < price:
            print("Sorry, you need to pay {0} dollars more. Please, pay properly.".format(price-money))
            money = 0
            continue
        if money == price:
            print("Thank you. Have a nice day!\n")
            print("=========Ready to receive new order.=========\n")
            break
        if money > price:
            print("Thank you. You gave me {0} dollars. Your change is {1} dollars.".format(money,money-price))
            print("\n=========Ready to receive new order.=========\n")
            break
```
