```python
import copy

class Student:
  Number = 0
  Name = ""
  sci = 0
  eng = 0
  mat = 0
  Sum = 0
  Avr = 0.0
  Grade = ""

  def __init__(self,num):
    self.Number = num

  def SetData(self):
    self.Name = input("Student Name : ")
    self.sci = int(input("Science Score : "))
    self.eng = int(input("English Score : "))
    self.mat = int(input("Math Score : "))
    self.Sum = self.sci + self.eng + self.mat
    self.Avr = self.Sum / 3.0
    if self.Avr >= 90:
      self.Grade = "A"
    elif self.Avr >= 80:
      self.Grade = "B"
    elif self.Avr >= 70:
      self.Grade = "C"
    elif self.Avr >= 60:
      self.Grade = "D"
    else:
      self.Grade = "F"

  def Show(self):
    print("==========={0}(st,nd,rd,th) student===========".format(self.Number))
    print("Student Name = {0}".format(self.Name))
    print("Science Score = {0}".format(self.sci))
    print("English Score = {0}".format(self.eng))
    print("Math Score = {0}".format(self.mat))
    print("Summation = {0}".format(self.Sum))
    print("Average = {0}".format(self.Avr))
    print("Grade = {0}".format(self.Grade))
    print("==============================\n")

  def Find(self,Name):
    if self.Name == Name:
      return True
    else:
      return False

class school:
    LowClass = []
    HighClass = []
    LowCount = 0
    HighCount = 0
    def SetStudent(self):
        Class = input("Set the class of student(Low/High): ")
        if Class == "Low" or Class == "low":
            self.LowCount += 1
            st = Student(self.LowCount)
            st.SetData()
            self.LowClass.append(copy.deepcopy(st))
        elif Class == "High" or Class == "high":
            self.HighCount += 1
            st = Student(self.HighCount)
            st.SetData()
            self.HighClass.append(copy.deepcopy(st))
        del st

    def ShowStudent(self):
        print("\n======Low Class======")
        for i in self.LowClass:
            i.Show()
        print("======High Class======")
        for i in self.HighClass:
            i.Show()

sc = school()
while 1:
    print("\n========Student Registering Program=========")
    print("        1.Register Student")
    print("        2.Show Registered Student Information")
    print("==============================================")
    choice = int(input("Choose your choice : "))
    if choice == 1:
        sc.SetStudent()
    elif choice == 2:
        sc.ShowStudent()
    
```
