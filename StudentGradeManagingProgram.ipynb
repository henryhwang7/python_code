```python
import copy
import os
number = 0
student = {}
school = []
cnnumber = 0
while 1:
  os.system('cls')
  print("{0:=^22}".format("Student's Grade Managing Program"))
  print("      1.Student Register")
  print("      2.Show Student information")
  print("      3.Search(student name,number)")
  print("      4.Show info based on grades")
  print("      5.Delete Student Information")
  print("      6.Save Student Information")
  print("      7.Import Student Information")
  print("      8.Quit")
  print("\n==== Number of registered students: {0} ====".format(cnnumber))
  choice = int(input("Choose your choice : "))
  if choice == 1:
      number += 1
      student["Number"] = number
      student["Name"] = input("Name : ")
      student["Gender"] = input("Gender : ")
      student["Age"] = input("Age : ")
      student["ScienceScore"] = int(input("Science Score : "))
      student["MathScore"] = int(input("Math Score : "))
      student["EnglishScore"] = int(input("English Score : "))
      student["Summation"] = student["ScienceScore"] + student["MathScore"] + student["EnglishScore"]
      student["Average"] = student["Summation"] / 3.0
      if student["Average"] >= 90:student["Grade"] = "A"
      elif student["Average"] >= 80:student["Grade"] = "B"
      elif student["Average"] >= 70:student["Grade"] = "C"
      elif student["Average"] >= 60:student["Grade"] = "D"
      else:student["Average"] = "F"
      cnnumber += 1
      school.append(copy.deepcopy(student))
      print("")
    
  elif choice == 2:
      for st in school:
          st_key = list(st.keys())
          print("========={0}st(th) student========".format(st["Number"]))
          for key in st_key:
              if key == "Number":
                  pass
              else:print("{0} : {1}".format(key,st[key]))
          print("==========================")
      input()
    
  elif choice == 3:
      index = input("Choose a way to search(Name,Number): ")
      index2 = input("Type {0} to search: ".format(index))
      for st in school:
          if str(st[index]) == index2:
              st_key = list(st.keys())
              print("========={0}(st,nd,rd,th) student========".format(st["Number"]))
              for key in st_key:
                  if key == "Number":
                      pass
                  else:print("{0} : {1}".format(key,st[key]))
              break
      input()
    
  elif choice == 4:
          Type = ["A","B","C","D","F"]
          for tp in Type:
                  print("======Grade {0}======".format(tp))
                  for st in school:
                          if st["Grade"] == tp:
                                  sy_key = list(st.keys())
                                  print("========={0}st(th) student========".format(st["Number"]))
                                  for key in st_key:
                                          if key == "Number":
                                                  pass
                                          else:print("{0} : {1}".format(key,st[key]))
                                  print("==========================")
          input()
        
  elif choice == 5:
          a = int(input("Type student number you want to delete : "))
          count = 0
          for st in school:
                  if st["Number"] == a:
                          del school[count]
                          cnnumber -= 1
                          break
                  count += 1
          input()
        
  elif choice == 6:
    while 1:
      st = input("Name your saving file : ")
      st += ".txt"
      try:
        f = open(st,'r')
      except:
        f = open(st,'w')
        break
      else:
        print("The file is already existed. Do you want to overwrite?(y,n): ")
        x = input()
        if x == 'y' or x == 'Y':
          f = open(st,'w')
          break
        elif x == 'n' or x == 'N':
          pass
    for st in school:
      st_key = list(st.keys())
      for key in st_key:
        f.write(str(st[key]) + '\n')
    f.close()
    
  elif choice == 7:
    x = input("Type name of the file you want to import : ")
    x += ".txt"
    try:
      f = open(x,'r')
    except:
      print("There is no such file.")
    else:
      temp = {}
      ct = 0
      cnnumber = 0
      lines = f.readlines()
      school = []
      key = ["Number","Name","Gender","Age","Science","Math","English","Summation","Average","Grade"]
      for line in lines:
        temp[key[ct]] = line[:-1]
        ct+=1
        if ct == 10:
          ct = 0
          cnnumber += 1
          school.append(copy.deepcopy(temp))
          temp = {}
            
  elif choice == 8:
          a = input("Do you want to terminate the program?(y,n)")
          if a == 'y' or a == 'Y':
                break
          elif a == 'n' or a == 'N':
                continue
```
